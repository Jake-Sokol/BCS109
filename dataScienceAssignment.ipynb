#The difference between lists and tuples

Lists are mutable, meaning their data can be changed.  Tuples on the other hand are immutable, meaning that their data cannot be changed.

Lists have a variable length, allowing you to change the size at runtime, while tuples have a fixed length, meaning that their size cannot be changed after allocation.

A benefit of using tuples is that operations with tuples can be executed faster than operations performed on lists.

#Python Libraries

**Scikit learn** - A library that allows for machine learning in Python. Machine learning considers a set of n samples of data and tries to predict properties of unknown data.

**Pandas** - The Python Data Analysis Library, which provides easy to use, high performance data structures.  It is used for data analysis and manipulation.

**Tensorflow** - A library for numerical computation and machine learning.  

**NumPy** - NumPy, or Numerical Python, is a library that is used for arrays.  NumPy provides an array object that is up to 50x faster than a Python list.
